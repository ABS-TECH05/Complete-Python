# Pandas-DataFrame And Series:
## Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). ðŸ¼
## ***i.e., Pandas lets you work with Excel-like data tables in Python for cleaning, filtering, and analysis.***

```python
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print(f'Series :\n {series}')
print(type(series))
```

```python
## Creating a Series from a dictionary
data = {'a':1, 'b':2, 'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index = ['a', 'b', 'c']
pd.Series(data, index=index)
```

```python
## DataFrame
## create a DataFrame from a dictionary of list

data = {
    'Name':['Surya', 'Krish', 'John'],
    'Age':[25, 30, 45],
    'City':["Bangalore", "New York", 'Florida']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## create a DataFrame from a list of dictionaries

data = [
    {'Name': 'Surya', "Age":25, 'City':'Bangalore'},
    {'Name': 'Krish', "Age":30, 'City':'New York'},
    {'Name': 'John', "Age":45, 'City':'Florida'}
]
df = pd.DataFrame(data)
print(df)
print(type(df))
```
