### Assignment 1: Array Creation and Manipulation

```python
# 1. Create a NumPy array of shape (5, 5) filled with random integers between 1 and 20. 
import numpy as np
a1 = np.random.randint(1, 21, size=(5,5))
print(f'Original array 1:\n{a1}\n')

## Replace all the elements in third column with 1.
a1[:, 2] = 1
print(f'Modified array 1:\n{a1}\n\n')
```

```python
# 2. Create a NumPy array of shape (4, 4) with values from 1 to 16.
a2 = np.arange(1,17).reshape(4,4)
print(f'Original array 2:\n{a2}\n')

## Replace the diagonal elements with 0.
np.fill_diagonal(a2, 0)
print(f'Modified array 2:\n{a2}')
```

### Assignment 2: Array Indexing and Slicing

```python
# 1. Create a NumPy array of shape (6, 6) with values from 1 to 36.
a1 = np.arange(1,37).reshape(6,6)
print(f'Original Array 1:\n{a1}\n')

## Extract the sub-array consisting of the 3rd to 5th rows and 2nd to 4th columns.
sub_a1 = a1[2:5, 1:4]
print(f'Modified Array 1:\n{sub_a1}\n\n')
```

```python
# 2. Create a NumPy array of shape (5, 5) with random integers.
a2 = np.random.randint(1, 21, size=(5,5))
print(f'Original Array 2:\n{a2}\n')

## Extract the elements on the border.
border_a2 = np.concatenate((a2[0,:], a2[-1, :], a2[1:-1, 0], a2[1:-1, -1]))
print(f'Border Elements:\n{border_a2}')
```

### Assignment 3: Array Operations

```python
# 1. Create two NumPy arrays of shape (3, 4) filled with random integers.
a1 = np.random.randint(1,11,size=(3,4))
a2 = np.random.randint(1,11,size=(3,4))
print(f'Original Array 1:\n{a1}')
print(f'\nOriginal Array 2:\n{a2}\n')

## Perform element-wise addition, subtraction, multiplication, and division.
print(f'Addition:\n{a1+a2}\n')
print(f'Subtraction:\n{a1-a2}\n')
print(f'Multiplication:\n{a1*a2}\n')
print(f'Division:\n{a1/a2}')
```

### Assignment 4: Statistical Operations

```python
# 1. Create a NumPy array of shape (5, 5) filled with random integers.
a = np.random.randint(1, 21, size=(5,5))
print(f'Original Array:\n{a}')

## Compute the mean, median, standard deviation, and variance of the array.
print(f'\nMean : {np.mean(a)}')
print(f'Median : {np.median(a)}')
print(f'Standard Deviation : {np.std(a)}')
print(f'Variance : {np.var(a)}')
```

```python
# 2. Create a NumPy array of shape (3, 3) with values from 1 to 9.
a = np.arange(1,10).reshape(3,3)
print('Array :\n',a)

## Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).
normalised_a = (a - np.mean(a))/np.std(a)
print('\nNormalized Array :\n', normalised_a)
```

### Assignment 5: Broadcasting

```python
# 1. Create a NumPy array of shape (3, 3) filled with random integers.
a = np.random.randint(1,11, size=(3,3))
print(f'Original Array:\n{a}\n')

## Add a 1D array of shape (3,) to each row of the 2D array using broadcasting.
r_a = np.random.randint(1,11, size=(3,))
print(f'1D Array :\n{r_a}\n')
print(f'Resulting Array :\n{a+r_a}')
```

```python
# 2. Create a NumPy array of shape (4, 4) filled with random integers.
a = np.random.randint(1,11, size=(4,4))
print(f'Array :\n{a}\n')

## Subtract a 1D array of shape (4,) from each column of the 2D array using broadcasting.
c_a = np.random.randint(1,11, size=(4,))
print(f'1D Array :\n{c_a}\n')
print(f'Resulting Array :\n{a - c_a[:, np.newaxis]}')
```

### Assignment 6: Linear Algebra

```python
# 1. Create a NumPy array of shape (3, 3) representing a matrix.
a = np.random.randint(1, 10, size=(3,3))
print(f'Matrix :\n{a}')

## Compute its determinant, inverse, and eigenvalues.
determinant = np.linalg.det(a)
inverse = np.linalg.inv(a)
eigen = np.linalg.eigvals(a)

print(f'\nDeterminant : {determinant}\n')
print(f'Inverse :\n {inverse}\n')
print(f'Eigen Values : {eigen}')
```

### Assignment 7: Advanced Array Manipulation

### Assignment 8: Fancy Indexing and Boolean Indexing

### Assignment 9: Structured Arrays

```python
# 1. Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float).
data_type = [('name', 'U10'), ('age', 'i4'), ('weight', 'f4')]
data = np.array([('Alice', 25, 55.5), ('Bob', 30, 85.3), ('Charlie', 20, 65.2)], dtype=data_type)
print(f'Array :  {data}')

## Add some data and sort the array by age.
sorted_data = np.sort(data, order='age')
print(f'Array sorted by age:  {sorted_data}')
```

```python
# 2. Create a structured array with fields 'x' and 'y' (both integers). 
data_type = [('x', 'i4'), ('y', 'i4')]
data = np.array([(1, 2), (3, 4), (5, 6)], dtype=data_type)
print(f'Array : {data}')

## Add some data and compute the Euclidean distance between each pair of points.
distances = np.sqrt((data['x'][:, np.newaxis]-data['x'])**2 - (data['y'][:, np.newaxis]-data['y']**2))
print(f'Euclidean Distances :\n{distances}')
```

### Assignment 10: Masked Arrays

```python
# 1. Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10.
import numpy.ma as ma
a = np.random.randint(1,21, size=(4,4))
masked_a = ma.masked_greater(a, 10)
print(f'Array :\n{a}')
print(f'Masked array :\n{masked_a}')

## Compute the sum of the unmasked elements.
print(f'\nSum of masked array : {masked_a.sum()}')
```

```python
# 2. Create a masked array of shape (3, 3) with random integers and mask the diagonal elements.
a = np.random.randint(1,21,size=(3,3))
masked_a = ma.masked_array(a, mask=np.eye(3, dtype=bool))
print(f'Array :\n{a}\n')
print(f'Diagonally masked array :\n{masked_a}\n')

## Replace the masked elements with the mean of the unmasked elements.
print(f'Modified masked array :\n{masked_a.filled(masked_a.mean())}')
```
