```python
n = int(input("Enter a number: "))
fact = 1
i = 1
while i <= n:
    fact *= i
    i += 1
print(f"The factorial of {n} is {fact}")
```

```python
n = int(input("Enter a number: "))
sum = 1
i = 1
while i <= n:
    sum += i
    i += 1
print(f"The sum of first {n} natural numbers is {sum}")
```
