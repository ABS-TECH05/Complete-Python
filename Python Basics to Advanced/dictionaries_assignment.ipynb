```python
d = {i:i**2 for i in range(1, 11)}
new_d = {k:v for k, v in d.items() if k % 2 == 0}
print(new_d)
```

```python
d = {i:i**2 for i in range(1, 6)}
print(f"Before swapping: {d}")

swapped_d = {v:k for k, v in d.items()}
print(f"After swapping: {swapped_d}")
```

```python
from collections import defaultdict
d = defaultdict(list)
d['a'].append(1)
d['b'].append(2)
d['c'].append(3)
print(dict(d))
```

```python
def count_char(s):
    flag = {}
    for i in s:
        if i in flag:
            flag[i] += 1
        else:
            flag[i] = 1
    return flag

string = "Hello World!"
print(count_char(string))
```
