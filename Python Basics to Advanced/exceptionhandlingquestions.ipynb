```python
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError as e:
        print(f"Error: {e}")
        result = None
    finally:
        print("Execution complete.")
    return result

print(divide(10, 2)) 
print(divide(10, 0))  
```

```python
def read_file(filename):
    try:
        file = open(filename, 'r')
        return file.read()
    except FileNotFoundError as e:
        print(f'Error: {e}')
    finally:
        try:
            file.close()
        except Exception as e:
            pass
            
read_file('data.txt')
```

```python
def add_lst(lst):
    try:
        result = sum(lst)
    except TypeError as e:
        print(f'Error : {e}')
        result = None
    finally:
        print('Execution Complete.')
    return result

print(add_lst([1, 2, 3, 'a']))
print(add_lst([1, 2, 3, 4]))
```

```python
def get_integer():
    try:
        value = int(input("Enter a integer."))
    except Exception as e:
        value = None
        print(f'Error: {e}')
    finally:
        print('Complete execution.')
    return value

print(get_integer())
```

```python
def get_dict_value(d, key):
    try:
        value = d[key]
    except Exception as e:
        print(f'Error: {e}')
        value = None
    finally:
        print('Execution complete.')
    return value
    
d = {'a': 1, 'b': 2, 'c': 3}
print(get_dict_value(d, 'b'))
print(get_dict_value(d, 'x'))
```

```python
def nested_exception(s):
    try:
        try:
            num = int(s)
        except Exception as e:
            print(f'Conversion Error: {e}')
            num = None
        finally:
            print('Conversion attempt complete.')
        if num != None:
            try:
                result = 10/num
            except Exception as e:
                print(f'Error : {e}')
                result = None
            finally:
                print('Division attempt complete.')
            return result
    finally:
        print('Overall execution complete.')
        
print(nested_exception('0'))
print(nested_exception('a'))
print(nested_exception('2'))     
```

```python
def get_list_element(lst, index):
    try:
        element = lst[index]
        return element
    except Exception as e:
        print(f"Error: {e}")
        pass
    finally:
        print('Execution complete.')

lst = [1,2,3,4,5]
print(get_list_element(lst, 2))
print(get_list_element(lst, 10))
```

```python
import requests

def read_url(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except Exception as e:
        print(f'Network Error: {e}')
        return None
    finally:
        print('Execution Complete.')
        
print(read_url('https://jsonplaceholder.typicode.com/posts/1'))
print(read_url('https://nonexistent.url'))        
```

```python
import json

def parse_json(str):
    try:
        data = json.loads(str)
        return data
    except Exception as e:
        print(f'JSON error : {e}')
        pass
    finally:
        print('Execution complete.')
        
print(parse_json('{"name": "John", "age": 30}'))
print(parse_json('Invalid JSON')) 
```

```python
class NegativeNumberError(Exception):
    pass

def check_for_negatives(lst):
    try:
        for i in lst:
            if i < 0:
                raise NegativeNumberError(f"Negative number found: {i}")
    except NegativeNumberError as e:
        print(f"Error: {e}")
    finally:
        print("Execution complete.")
        
check_for_negatives([1, -2, 3, 4])
check_for_negatives([1, 2, 3, 4])
```

```python
def risky():
    raise ValueError('An Error has occured in risky function.')

def safe():
    try:
        risky()
    except ValueError as e:
        print(f'Error: {e}')
    finally:
        print('Execution complete.')
        
safe()
```

```python
class Calculator:
    def divide(self, a, b):
        try:
            result = a / b
        except ZeroDivisionError as e:
            print(f"Error: {e}")
            result = None
        finally:
            print("Execution complete.")
        return result
    
calc = Calculator()
print(calc.divide(10, 2))
print(calc.divide(10, 0))  
```

```python
def convert_to_integers(lst):
    integers = []
    try:
        for item in lst:
            integers.append(int(item))
    except ValueError as e:
        print(f"Error: {e}")
        integers = None
    finally:
        print("Execution complete.")
    return integers

print(convert_to_integers(['1', '2', 'three', '4']))  
print(convert_to_integers(['1', '2', '3', '4']))
```

```python
def convert_with_comprehension(lst):
    try:
        integers = [int(i) for i in lst]
    except ValueError as e:
        print(f"Error: {e}")
        integers = None
    finally:
        print("Execution complete.")
    return integers
        
print(convert_with_comprehension(['1', '2', 'three', '4']))
print(convert_with_comprehension(['1', '2', '3', '4']))
```

```python
def write_to_file(lst, filename):
    try:
        file = open(filename, 'w')
        for i in lst:
            file.write(i, '\n')
    except Exception as e:
        print(f'Error: {e}')
    finally:
        try:
            file.close()
        except Exception:
            pass
        
write_to_file(['Hello', 'World'], 'output.txt')
```
