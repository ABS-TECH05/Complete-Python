```python
import random
random_lst = [random.randint(1, 10) for i in range(15)]
print(f"Original list: {random_lst}")

uniq_lst = list(set(random_lst))
asc_lst = sorted(random_lst)
desc_lst = sorted(random_lst, reverse=True)

print(f"Unique elements: {uniq_lst}")
print(f"Ascending order: {asc_lst}")
print(f"Descending order: {desc_lst}")
```

```python
students = [
    {'name': 'Alice', 'score': 88},
    {'name': 'Bob', 'score': 72},
    {'name': 'Charlie', 'score': 95},
    {'name': 'David', 'score': 65},
    {'name': 'Eve', 'score': 78}
]
print("Original list of students:")
for i in students:
       print(i)

print("Students sorted by score (descending):")    
sorted_students = sorted(students, key=lambda x: x["score"], reverse=True)
for i in sorted_students:
    print(i)
```

```python
lst = [[1, 3, 5],
     [5, 8, 4],
     [9, 0, 2]]
print("Original list: ")
for i in lst:
    print(i)
        
flattened_lst = [i for j in lst for i in j]
print(f"Flattened list: {flattened_lst}")
```

```python
def rotate_list(l, n):
    return l[n:] + l[:n]

lst = [i for i in range(1, 6)]
print(f"Original list: {lst}")

rot_lst = rotate_list(lst, 2)
print(f"List after rotating by 2 positions: {rot_lst}")
```

```python
import random
lst1 = [random.randint(1, 12) for i in range(10)]
lst2 = [random.randint(1, 12) for i in range(10)]

inner_join = list(set(lst1) & set(lst2))
print(f"List 1: {lst1}")
print(f"List 2: {lst2}")
print(f"Inner join (common elements): {inner_join}")
```
