```python
s1 = {i for i in range(1,6)}
s2 = {i for i in range(2, 10, 2)}
print(f"Union: {s1|s2}")
print(f"Intersection: {s1 & s2}")
print(f"Difference (s1 - s2): {s1 - s2}")
print(f"Symmetric Difference: {s1 ^ s2}")
```
