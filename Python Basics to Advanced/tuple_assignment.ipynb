```python
def max_in_tpl(tpl):
    return max(tpl)

def min_in_tpl(tpl):
    return min(tpl)

def sum_of_tpl(tpl):
    return sum(tpl)

sample_tpl = (3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5)
print(f"Max: {max_in_tpl(sample_tpl)}")
print(f"Min: {min_in_tpl(sample_tpl)}")
print(f"Sum: {sum_of_tpl(sample_tpl)}")
```
